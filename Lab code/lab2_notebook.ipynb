# 1.  ASCII and int<->character conversion

[ASCII](https://en.wikipedia.org/wiki/ASCII) abbreviated from American Standard Code for Information Interchange, is a character encoding standard for electronic communication. The first edition of the standard was published in 1963.

There are other character encoding standards like UTF-8, which supports many non-latin languages like Arabic, Chinese, Cyrillic, Devanagari, Greek, Hebrew, Japanese, Korean, Tamil and Thai script. It even includes [emoji](https://unicode.org/emoji/charts-14.0/full-emoji-list.html).

```python
'\U0001f600', '\U0001f44d'
```

For now, let's focus on ASCII encoding. Each English character have a corrsponding integer, in the range of 0-127.  
For example:  
A<->65, B<->66, ..., Z<->90.  Uppercase letters are in the range `[65, 90]`  
a<->97, b<->98, ..., z<->122. Lowercase letters are in the range `[97, 122]`  
See the full table [here](https://theasciicode.com.ar/ascii-printable-characters/ampersand-ascii-code-38.html)


Luckily, we don't have to memorize this table. [`ord()`](https://docs.python.org/3/library/functions.html#ord) and [`chr()`](https://docs.python.org/3/library/functions.html#chr) will do the conversion for you.

In Python, there is no `char` type. We use a `string` with 1 element to represent character.

```python
ord('A'), ord('a'), chr(65), chr(97)
```

```python
x = "Hello From Python"

# convert string to a list of characters
print(list(x))

# split string into a list of tokens
tokens = x.split(' ')
print(tokens)

# join a list of token back to string
# split and join is the inverse operation of each other
print(','.join(tokens))
print(' '.join(x.split(' '))) # same as x

```

1. Write a function to swap the case of input string. aka, change an uppercase character to lowercase, lowercase character to uppercase, keep non a-z character unchanged.

```python
'5aBCd' - > '5AbcD'
```

```python
def swap_case(s: str):
    buffer = []
    for c in s:
        if c.islower():
            c = c.upper()
        elif c.isupper():
            c = c.lower()
        buffer.append(c)
    return ''.join(buffer)
swap_case('5aBCd')
```

# 4. Work with File.
### Read File

```python
# document of open() function: https://docs.python.org/3/library/functions.html#open

# 1st argument: filename
# 2nd argument: mode. Common modes: `r`, `w`, `rb`, `wb`.
# With `b` suffix means it is opend in binrary mode.
# Without `b` suffix means it is opened in text mode, i.e. line by line.
fd = open("WeatherData.txt", "r") # fd is a File object. You need call functions to read from it to access its content
print(fd)

# read one line
line = fd.readline()
print(line)
print(repr(line)) # Use repr to print string literal, show escape characters like the line ending

# read all lines
fd = open("WeatherData.txt", "r") # We need to open it again because the file has been partially read
lines = fd.readlines() # read the entire file into a list of lines
print(len(lines))      # the total number of lines
print(repr(lines[0]))  # the first line

# read all into a string
fd = open("WeatherData.txt", "r") # We need to open it again because the file has been read to the end
all_content = fd.read() # a very long string store all of its content
print(len(all_content)) # the total number of characters

# iterate over file object using for loop
fd = open("WeatherData.txt", "r") # We need to open it again because the file has been read to the end
for line in fd:
    print(line)
```

# 5. Read / Write CSV

[CSV](https://en.wikipedia.org/wiki/Comma-separated_values) file stands for comma/space/tab separated values.

CSV files are organized in two levels.
1. Entire file is organized line by line, from top to bottom.
2. Each line is organized token by token, from left to right.

Therefore, we need two levels of loop to read a CSV file.
1. Loop over all the lines  
2. Inside each line, loop over all the tokens  


```python
# list comprehension
with open("WeatherData.txt") as fin:
    for line in fin: # iterate over file object using for loop
        tokens = line.rstrip().split("\t") # each item is a str
        print(repr(line))
        print(tokens)
        tokens_float = [float(x) for x in tokens]
        '''
        tokens_float = []
        for x in tokens:
            x_float = float(x)
            tokens_float.append(x_float)
        '''
        print(tokens_float)
```

### Do some processing and write to another file

```python
with open("WeatherData.txt", "r") as fin:
    with open("WeatherData_out.txt", "w") as fout:
        for line in fin:
            tokens = line.rstrip().split("\t") # each item is a str
            tokens_float = [float(x) for x in tokens] # each item is a float
            tokens_float2 = [x * 2 + 10 for x in tokens_float] # multiply each item by 2
            line_processed = '\t'.join([str(x) for x in tokens_float2])
            fout.write(line_processed  + '\n')
```
