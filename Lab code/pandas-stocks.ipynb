# Pandas Example: Stock Market
### Using _Yahoo! Finance_ library

```python
import pandas as pd
#Data Source
import yfinance as yf
```

```python
aapl = yf.download(tickers='AAPL', period='100d', interval='1d')  # Apple
cvx  = yf.download(tickers='CVX',  period='100d', interval='1d')  # Chevron
hon  = yf.download(tickers='HON',  period='100d', interval='1d')  # Honeywell
wmt  = yf.download(tickers='WMT',  period='100d', interval='1d')  # Walmart
```

```python
#Print data
aapl
```

```python
# Plot data
aapl[['High', 'Low']].plot()
```

### Correlation analysis
__Correlation__ is a statistical term describing the degree to which two variables move in coordination with one another. If the two variables move in the same direction, then those variables are said to have a positive correlation. If they move in opposite directions, then they have a negative correlation.
Values can range from -1 to +1. A correlation of 1.0 indicates that two variables are perfectly coordinated. A correation of -1.0 indicates that the variables are also perfectly coordinated but in opposite directions, e.g. higher values of one variable correspond to lower values for the other variable. A correlation around 0.0 indicates that the variables are independent, i.e. that there is no statistical connection. 

```python
aapl.corr()
```

```python
# Combine data
stocks = pd.DataFrame({'Apple': aapl['Close'], 'Chevron': cvx['Close'], 
                       'Honeywell': hon['Close'], 'Walmart': wmt['Close']}); stocks
```

### Visualization with Pandas' own plot method

```python
stocks.plot()
```

```python
stocks.corr()
```

## Repeating the above for scaled-up data: 5000 business days (up from 100)

```python
aapl = yf.download(tickers='AAPL', period='5000d', interval='1d')  # Apple
cvx  = yf.download(tickers='CVX',  period='5000d', interval='1d')  # Chevron
hon  = yf.download(tickers='HON',  period='5000d', interval='1d')  # Honeywell
wmt  = yf.download(tickers='WMT',  period='4997d', interval='1d')  # Walmart
```

```python
stocks = pd.DataFrame({'Apple': aapl['Close'], 'Chevron': cvx['Close'], 
                       'Honeywell': hon['Close'], 'Walmart': wmt['Close']}); stocks
```

```python
stocks.plot()
```

```python
stocks.corr()
```

### Conclusion
Over a longer time frame (5000 business days), 
the above stocks correlate much more than over a shorter time frame (100 business days).
